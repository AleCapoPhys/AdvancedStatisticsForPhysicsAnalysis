# Exercise 1 - vectors and data frames
Create vectors, holding the lake’s name and all the parameters and build a dataframe
called scottish.lakes from the vectors.

```r
Loch <- c('Loch Ness','Loch Lomond','Loch Morar','Loch Tay','Loch Awe',
           'Loch Maree','Loch Ericht','Loch Lochy','Loch Rannoch',
           'Loch Shiel','Loch Katrine','Loch Arkaig','Loch Shin')
Volume <- c(7.45, 2.6, 2.3, 1.6, 1.2, 1.09, 1.08, 1.07, 0.97, 0.79, 0.77, 0.75, 0.35)
Area <- c(56, 71, 27, 26.4, 39, 28.6, 18.6, 16, 19, 19.5, 12.4, 16, 22.5)
Length <- c(39, 36, 18.8, 23, 41, 20, 23, 16, 15.7, 28, 12.9, 19.3, 27.8)
Max_depth <- c(230, 190, 310, 150, 94, 114, 156, 162, 134, 128, 151, 109, 49)
Mean_depth <- c(132, 37, 87, 60.6, 32, 38, 57.6, 70, 51, 40, 43.4, 46.5, 15.5)
```

```r
scottish.lakes <- data.frame(Loch, Volume, Area, Length, Max_depth, Mean_depth)
```

```r
scottish.lakes
```

### 1 - evaluate the highest and lowest volume and area lake

This was accomplished using the functions **which.min()** and **which.max()** on the Volume and Area columns. These functions return the index of the row in which the respective min or max is.
Said index was then used to access the dataframe, which was displayed together with the lake's other attributes, by using the **","**  .

```r
cat('Min and max volume lakes are respectively:')
scottish.lakes[which.min(scottish.lakes$Volume), ]
scottish.lakes[which.max(scottish.lakes$Volume), ]
cat('Min and max area lakes are respectively:')
scottish.lakes[which.min(scottish.lakes$Area), ]
scottish.lakes[which.max(scottish.lakes$Area), ]
```

### 2  - order the frame with respect to the area and determine the two largest area lakes

This was accomplished using the **order()** function in R. 
The two largest area lakes are then determined by "slicing" the resulting dataframe.

```r
scottish.lakes[order(Area, decreasing = TRUE),]
scottish.lakes[order(Area, decreasing = TRUE),][1:2,]
```

### 3 - by summing up the areas occpupied by the lakes, determine the area of Scotland covered by water

This was accomplished using the **sum()** function in R and passing as the argument the Area column of the dataframe.

```r
covered_in_water <- sum(scottish.lakes$Area)

cat(paste("The area of scotland which is covered in water is", covered_in_water, "km^2"))
```

# Exercise 2 - Crude Oil Production

### 1- Write R code that is able to read the file and import it in a data frame structure. (Hint: before loading the file, open it with a text editor and check its structure).

This was accomplished using the **read.csv()** function in R. For ease of operations then the last column of the dataframe was renamed to "Price". 

```r
crude_oil_prices <- read.csv("crude-oil-prices.csv", sep = ',')
head(crude_oil_prices)
```

```r
names(crude_oil_prices)[4] <- "Price"
head(crude_oil_prices)
```

### 2 - produce a plot with the Oil price as a function of the year

This was accomplished with the help of the **ggplot2** library.

```r
library("ggplot2")

ggplot(crude_oil_prices, aes(x = Year, y = Price)) +
    geom_point(color = 'red') +
    geom_line(color = 'black') +
    scale_x_continuous(name="Year", breaks=seq(min(crude_oil_prices$Year)-1,max(crude_oil_prices$Year)+1, 10)) +
    ggtitle('Crude Oil Prices, in $/Barrel')+ 
    theme(plot.title = element_text(hjust = 0.5))
```

### 3 - which is the highest price in history ? When did it occur ?
To find out I used the function **which.max()** on the dataset, specifying that I wanted the index of the maximum price, I then printed the whole row, thus showing the year in which it occured as well.

```r
crude_oil_prices[which.max(crude_oil_prices$Price), ]
```

### 3 - plot the derivative of the curve, simply evaluated with the finite difference formula (forward derivative).

In order to do this the **diff()** function in R was used. The diff function takes in a vector of elements and computes the difference between each element and the previous one. For example: <br>
**x <- c(5, 2, 10, 1, 3)** <br>
**diff(x) returns  (-3  8 -9  2)** <br>
As can be seen the resulting array will have lenght 1 less than the original one. So proper care will be taken for the derivative of the price in the last year available, which will be unknown. 

```r
library("tidyverse")

crude_oil_prices <-mutate(crude_oil_prices, Derivative = c(diff(crude_oil_prices[, 'Price']), NA))
head(crude_oil_prices)
```

```r
ggplot(crude_oil_prices, aes(x = Year, y = Derivative)) +
    geom_point(color = 'red') +
    geom_line(color = 'black') +
    scale_x_continuous(name="Year", breaks=seq(min(crude_oil_prices$Year)-1,max(crude_oil_prices$Year)+1, 10)) +
    ggtitle('Rate of change of Oil Prices throughout the years')+ 
    theme(plot.title = element_text(hjust = 0.5))
```

#  Exercise 3 - World Coal Production

### 1 - Write R code that is able to read the file and import it in a tibble structure

This was accomplished using read.csv and the function tibble.

```r
library("tibble")
```

```r
world_coal_prod <- tibble(read.csv("coal-production-by-country.csv", header = TRUE))
head(world_coal_prod)

names(world_coal_prod)[4] <- "Prod"

head(world_coal_prod)
```

### 2 - count the number of countries available in the file and produce a barplot with the number of entries for each country

Firstly the dataset had to be purged of combinations of countries, such as "Asia And Oceania" etc... (a lot of time was spent trying to think of a quick way to accomplish this, but at the end it was done "manually", by writing the names of the rows which aren't countries and applying a fiter).<br>
The rest was accomplished using the barplot function in ggplot. The aestethic **label = ..count..** displays the number of entries for each country. 

```r
combinations <- c('Africa','Asia and Oceania','Asia Pacific','Central and South America','CIS','EU-28',
             'Eurasia','Europe (other)','Europe','North America','OECD','OPEC','Other Africa','Other Asia & Pacific',
             'Other CIS', 'Other S. & Cent. America', 'South & Central America','USSR','World') 
coal_no_rep <- filter(world_coal_prod, !( Entity %in% combinations))
```

```r
#This cell has trouble running correctly on the first try. I wonder why.
#The first time it runs it doensn't ignore what I told it to ignore in the cell above, but then it does.

library("ggplot2")

ggplot(coal_no_rep, aes(x=Entity)) +
    geom_bar(color="cadetblue4", fill = "navy")+
    geom_text(stat = 'count',aes(label =..count..),hjust = 1.5,color="ivory4", size=4)+
    coord_flip()
options(repr.plot.width=5, repr.plot.height=30)
```

### selecting only the year after 1970, determine the total integrated production for each country and print the top 5 Countries with highest coal productions

Firstly all years prior to 1970 were removed with a filter.<br>
Then the total integrated production for each country was obtained by summing all the remanining available values for coal production for each country and by creating a new dataset. 


```r
library("tidyverse")

coal_after_1970 <- filter(coal_no_rep, Year>= 1970)

total_prod <- coal_after_1970 %>% group_by(Entity) %>% summarise(TotalCoal= sum(Prod))

ordered <- total_prod[order(total_prod$TotalCoal, decreasing = TRUE),]
ordered[1:5,]
```

### 4 - for the 5 top Countries, create a plot of production as a function of time

This was accomplished using a filter to select only the 5 most coal producing countries in the world, then by plotting the values of coal produced by them per year.

```r
biggest_prod <- filter(coal_after_1970,
                       Entity=='China'|
                       Entity=='United States'|
                       Entity=='Russia'|
                       Entity =='Australia'|
                       Entity =='India')
options(repr.plot.width=10, repr.plot.height=6)
ggplot(biggest_prod, aes(x = Year, y = Prod, col = Entity)) +
    geom_point(size = 1) +
    geom_line()+
    scale_y_continuous(name="Coal production [TW/h]")
```

### 5 - generate a plot with the cumulative sum of the World’s coal production over the years

```r
world <- filter(world_coal_prod, Entity=='World')
world <- mutate(world, CumulativeProd = cumsum(Prod))

ggplot(world, aes(x = Year, y = CumulativeProd)) +
    geom_point(size = 2, col = 'red') +
    geom_line(col='black') +
    scale_y_continuous(name="Cumulative production of Coal in the world [TW/h]")
```

The file vaccinationsby-manufacturer.csv, (https://github.com/owid/covid-19-data/blob/

### 1 - filter() the original tibble by selecting the following countries: Italy

Everything is standard here, but note that I converted the **date** column into the Date format. This will become important later, and also it is good practice.

```r
library("tibble")

vaccines <- tibble(read.csv("vaccinationsby-manufacturer.csv", header = TRUE))

head(vaccines)

vaccines_italy <- vaccines[vaccines$location=='Italy', ]
vaccines_italy$date <- as.Date(vaccines_italy$date)

head(vaccines_italy)
```

### 2 - plot the number of vaccines given as a function of time for the different vaccine manufacturer

```r
library("ggplot2")
library("scales")

options(repr.plot.width = 10, repr.plot.height = 6)

ggplot(vaccines_italy, aes(x = date, y = total_vaccinations, col = vaccine)) + 
geom_point(size = 1)+
scale_x_date(name = "Date",
                 breaks = function(x) seq.Date(from = min(x), to = max(x), by = "2 months"), 
                 labels = date_format("%d-%m-%Y"))
```

### 3 - from the same tibble plot the total number of vaccines shot per day in Italy

This was accomplished by grouping by date and summing the total_vaccinations column.

```r
italy_vaccinations_by_day <- vaccines_italy %>% group_by(date) %>% summarise(vaccines_by_day = sum(total_vaccinations))

library("ggplot2")
library("scales")

options(repr.plot.width = 10, repr.plot.height = 6)

ggplot(italy_vaccinations_by_day, aes(x = date, y = vaccines_by_day)) + 
geom_line(size = 0.5)+
scale_x_date(name = "Date",
                 breaks = function(x) seq.Date(from = min(x), to = max(x), by = "2 months"), 
                 labels = date_format("%d-%m-%Y"))+ 
    scale_y_continuous(name= "Vaccinations by day in Italy")
```
